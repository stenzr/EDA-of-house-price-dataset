### lets observe the correlation first

```python
#generating the correltion matrix

correlation_matrix = datatrain.corr()
correlation_matrix.head()
```

```python
# heatmap of correlation among various attributes

sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})
plt.subplots(figsize=(30,30))
p = sns.heatmap(correlation_matrix, annot=True, lw=1.5, fmt='.2f', cmap='coolwarm')
rotxlabel = p.set_xticklabels(p.get_xticklabels(),fontdict={'fontsize':20}, rotation=90)
rotylabel = p.set_yticklabels(p.get_yticklabels(),fontdict={'fontsize':20}, rotation=30)
```

### Following conclusions can be drawn :
* not many attributes are correlated with each other
* __SalePrice__ _the target variable_ is strongly related with only few attributes
* we can conclude the data set has a lot of attributes that can be dropped without hampering the overall impact


```python
#getting 10 strong correlated features and exploring relation with SalePrice

strongAttributes = correlation_matrix.nlargest(10, 'SalePrice')['SalePrice'].index
```

```python
strongAttributes
```

```python
cm = np.corrcoef(datatrain[strongAttributes].values.T)
plt.subplots(figsize=(15,15))
#sns.set(font_scale=1.25)
p1 = sns.heatmap(cm, cbar=True, annot=True,lw= 1.5, fmt='.2f',cmap='coolwarm', yticklabels=strongAttributes.values, xticklabels=strongAttributes.values)
rotxlabel1 = p1.set_xticklabels(p1.get_xticklabels(),fontdict={'fontsize':15}, rotation=90)
rotylabel1 = p1.set_yticklabels(p1.get_yticklabels(),fontdict={'fontsize':15}, rotation=30)
plt.show()
```
